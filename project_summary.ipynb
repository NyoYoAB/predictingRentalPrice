# Rental Price Estimation Project

This project focuses on estimating rental prices for homeowners, real estate investors, and property management companies. The simulator helps answer critical business questions such as:

- **Should I rent or sell a property?**
- **How much rental income can I expect?**
- **Is managing this property for rental income worthwhile?**

---

## Key Sections of the Presentation:

### About the Project:
The goal is to estimate the monthly rental income based on a property's value, using a predefined percentage (0.4% of the sale value). This estimation supports decision-making regarding property rental or sale.

---

### EDA and Feature Engineering:
Two key features were created to capture non-linear relationships:
- **inverse_house_age**: Accounts for non-linear relationships between the age of the house and its value.
- **year_sqrt**: Created by multiplying the house's year of construction with its square footage to capture potential interaction effects.

In total, 12 features were selected for modeling, and 10 were scaled using the MinMax method.

---

### Model Evaluation:
Several regression models were evaluated, including:
- **Simple Linear Regression**:
    - R-squared: 0.2669
    - MAPE: 18.3741
- **Fast Simple Linear Regression**:
    - R-squared: 0.512
    - MAPE: 15.6461
- **Multivariate Linear Regression**:
    - R-squared: 0.5082
    - MAPE: 15.1413
- **KNN Regressor (K=5)**:  
    - **R-squared**: 0.5987  
    - **MAPE**: 13.548  
    - *This model was chosen for the application due to its better accuracy and lower errors.*

---

### Summary:
- The **KNN Regressor model (K=5)** was selected as the optimal solution due to its higher R-squared value and lower Mean Absolute Percentage Error (MAPE).
- The app provides practical, fast, and accurate rental price estimates for **real estate investors** and **rental management**.
